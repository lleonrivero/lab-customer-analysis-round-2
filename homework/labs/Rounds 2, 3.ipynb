# Show the dataframe shape.

```python
import pandas as pd
dataset = pd.read_csv('marketing_customer_analysis.csv')
pd.set_option('display.max_columns', None) # Used to show all columns
display(dataset.head())
display(dataset.tail())
dataset.shape
```

# Standardize header names.

```python
dataset.columns
```

```python
# Standardizing header names to all lower and removing column 'Unnamed: 0'

cols = []
for i in range(len(dataset.columns)):
    cols.append(dataset.columns[i].lower())
dataset.columns = cols
dataset.columns

dataset = dataset.drop(['unnamed: 0'], axis=1)

dataset
```

# Which columns are numerical? Which columns are categorical?

```python
# Showing data types
# float64 and int64 are numerical
# object is categorical
dataset.dtypes
```
