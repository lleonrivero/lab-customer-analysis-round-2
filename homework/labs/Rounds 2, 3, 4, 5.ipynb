# Show the dataframe shape.

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('marketing_customer_analysis.csv')
pd.set_option('display.max_columns', None) # Used to show all columns
display(dataset.head())
display(dataset.tail())
dataset.shape
```

# Standardize header names.

```python
dataset.columns
```

```python
# Standardizing header names to all lower and removing column 'Unnamed: 0'

dataset = dataset.drop(['Unnamed: 0'], axis=1)

cols = []
for i in range(len(dataset.columns)):
    cols.append(dataset.columns[i].lower().replace(' ', '_'))
dataset.columns = cols
dataset.columns



dataset
```

# Which columns are numerical? Which columns are categorical?

```python
# Showing data types
# float64 and int64 are numerical
# object is categorical
dataset.dtypes
```

# Check and deal with NaN values.

```python
dataset = dataset.drop_duplicates()
dataset.shape # Only removed a few lines.
```

```python
dataset.isna().sum()
```

```python
round(dataset.isna().sum()/len(dataset),4)*100 # shows the percentage of null values in a column
```

```python
dataset['vehicle_type'].value_counts()
```

```python
# 1. All values in column 'vehicle type' are either A or NaN. Uninteresting.
# 2. In addition, column vehicle type is missing 50% of the values. Not worth it replacing NaN by A. Column to be removed.

dataset = dataset.drop(['vehicle_type'], axis=1)
dataset
```

```python
# Finally decided to remove NaN rows for 2 reasons:
    # 1. We are still keeping a large amount of data points. 
    # 2. When trying to run linear model, there were too many NaNs and it wouldn't run. 
dataset = dataset.dropna()
dataset.shape
dataset
```

```python
dataset['income'].value_counts()
# Manye values for income are 0... Understand why.
```
