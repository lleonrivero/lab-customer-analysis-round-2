# Reporting - Present results

Insurance company requests the creation of a model capable of predicting the claim amount based on the profiling of its customers. To do so, it provides a dataset with data from 10,910 customers, with a total of 24 explanatory variables and the target dependent variable ('total_claim_amount').

First, the data is audited for better understanding, and cleaned for further analysis. Cleaning includes the removal of a column ('vehicle_type') with an insufficient amount of data, and the deletion of repeated rows or rows with null values that are impossible to retrieve. The row deletion process reduces the data set to 9,154 data points, which, in any case, is considered sufficient to be able to perform the required tasks without reducing the quality of the result. 

Subsequently, the numerical values are separated from the categorical values, which makes it possible to normalize the numerical values using the MinMaxScaler tool, placing all their values between 0 and 1. 

To be able to work with the categorical variables estimated to have the greatest impact on the number of claims, the following are isolated: 'coverage', 'policy_type', 'policy' and 'vehicle_class'. These variables are transformed into numerical ones through the OneHotEncoder and LabelEncoder processes, using the most appropriate estimated criteria for each case. The next step consists in reconstructing the dataset by incorporating the transformed categorical variables into the original numerical variables. In this way, the training and testing of the model will be performed on a total of 26 explanatory variables. 

In the first stage, 80% of the data are used for training, and the remaining 20% for testing. The R2 of the training is 0.519 and that of the test is 0.527. In both cases, these values are at the limit of what we could qualify as a good linear regression model. Given these results, it is decided to process the data again, this time normalizing the numerical values using the StandardScaler tool, which should place their values between -3 and 3. The linear regression model is run again with this new normalization of the data, reducing the amount of data for training to 70%, and suppressing the variable 'vehicle_class'. The model, however, behaves virtually identically, yielding an R2 in the training of 0.515 and 0.527 in the test. To confirm these results, a third trial is performed: the first normalization is recovered, the 'policy' variables are also removed, and the training is rerun with 80% of the data. However, the results are repeated: training R2, 0.517; test R2, 0.525.

Conclusion:
The available data do not allow us to improve the model beyond an R2 of between 0.515 and 0.527. In these circumstances, where our results are at the minimum threshold of reliability, and once presented to the client, the recommendation is to request a new data set with more data points and new variables.
